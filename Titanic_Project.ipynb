```python
def male_female_child(passenger):                 # Function for seperating male,female and child
    age,sex=passenger
    
    if(age<16):
        return 'child'
    else:
        return sex
```

```python
df['person']=df[['Age','Sex']].apply(male_female_child,axis=1)   # Person column
```

```python
fig=sns.FacetGrid(df,hue='person',aspect=4)          # KDE Plot using FacetGrid
fig.map(sns.kdeplot,'Age',shade=True)
oldest=df['Age'].max()
fig.set(xlim=(0,oldest),ylim=(0,0.1))
fig.add_legend()
```

```python
fig=sns.FacetGrid(df,hue='Pclass',aspect=4)     # Let;s do it with pclass
fig.map(sns.kdeplot,'Age',shade=True,bw='silverman')
oldest=df['Age'].max()
fig.set(xlim=(0,oldest),ylim=(0,0.1))
fig.add_legend()
```

```python
fig=sns.FacetGrid(df,hue='person',aspect=4)   # Rugplot with facetgrid
fig.map(sns.rugplot,'Age',height=0.5)
oldest=df['Age'].max()
fig.set(xlim=(0,oldest),ylim=(0,0.1))
fig.add_legend()
```

```python
deck=df['Cabin'].dropna()
```

```python
deck.head()
```

```python
# Function for creating level object
levels=[]
for level in deck:
    levels.append(level[0])

cabin_df=DataFrame(levels)
cabin_df.columns=['Cabin']
sns.factorplot('Cabin',data=cabin_df,kind='count',palette='winter_d')   # Factorplot for deck levels
```

```python
# Redefine cabin_df as everything but where the row was equal to 'T'
cabin_df = cabin_df[cabin_df.Cabin != 'T']
#Replot
sns.factorplot('Cabin',data=cabin_df,kind='count',palette='summer')
```
